```python
import numpy as np
import matplotlib.pyplot as plt

# Given slope and intercept
slope = 2.3
intercept = 10

# Generate some time values
time_values = np.arange(0, 10, 1)  # Assuming time values from 0 to 9

# Calculate corresponding tree heights using the linear regression equation
tree_heights = slope * time_values + intercept

# Plot the graph
plt.plot(time_values, tree_heights, label='Linear Regression')

# Add labels and title
plt.xlabel('Time (years)')
plt.ylabel('Tree Height')
plt.title('Linear Regression: Tree Height Over Time')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
